```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
folder_dir = "img/"
imgs = []
for image in os.listdir(folder_dir):
    curImg = cv.imread(folder_dir + image)
    imgs.append((cv.cvtColor(curImg, cv.COLOR_BGR2GRAY)))
```

```python
for i,img in enumerate(imgs):
    plt.figure()
    plt.imshow(img, 'gray')
    plt.title(os.listdir(folder_dir)[i])
    plt.show()
    plt.figure()
    plt.hist(img.ravel(),256,[0,256])
    plt.title(os.listdir(folder_dir)[i])
    plt.show()
```

```python
thresh1 = 70
im_bw = cv.threshold(imgs[0], thresh1, 255, cv.THRESH_BINARY)[1]
plt.imshow(im_bw, 'gray')
```

```python
thresh2 = 100
im_bw = cv.threshold(imgs[1], thresh2, 255, cv.THRESH_BINARY)[1]
plt.imshow(im_bw, 'gray')
```

```python
thresh3 = 80
im_bw = cv.threshold(imgs[2], thresh3, 255, cv.THRESH_BINARY)[1]
plt.imshow(im_bw, 'gray')
```

```python
thresh4 = 50
im_bw = cv.threshold(imgs[3], thresh4, 255, cv.THRESH_BINARY)[1]
plt.imshow(im_bw, 'gray')
```

## Reflection

a. We selected our thresholds based on the histograms and tried to identify peaks and dips as different peaks correspond to different elements in the image. By picking the dip as a threshold, we can differentiate between different elements.

b. For images where different elements with similar color values (img 3 above) are adjacent to each other, it is difficult to perform segmentation as they will be grouped together as one peak in the histogram. For images where there is distinct separation between differnt elements, the binary conversion allows for easy segmentation.
